```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
import statsmodels .formula.api as smf
```

```python
data1=pd.read_csv("NewspaperData.csv")
data1
```

```python
data1.info()
```

```python

fig,axes=plt.subplots(2,1,figsize=(8,6),gridspec_kw={'height_ratios':[1,3]})
sns.boxplot(data=data1['Newspaper'],ax=axes[0],color='skyblue',width=0.5,orient='h')
axes[0].set_title('Boxplot')
axes[0].set_xlabel(' Sales')
sns.histplot(data1["daily"],kde=True,ax=axes[1],color='purple',bins=30)
axes[1].set_title("Newspaper")
axes[1].set_xlabel("Daily")
axes[1].set_ylabel("Sunday")
plt.tight_layout()
plt.show()
```
